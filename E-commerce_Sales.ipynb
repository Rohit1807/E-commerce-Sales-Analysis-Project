<span style="color:red; font-size:15px">1. Sales Analysis</span>


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# Total Sales by Product Type and Location
sales_by_product =  ecommerce_data.groupby('Product_Type')['Total_Sales'].sum()
sales_by_location =  ecommerce_data.groupby('CustomerLocation')['Total_Sales'].sum()

# Plotting Sales by Product Type
plt.figure(figsize=(10, 7))
sns.barplot(x=sales_by_product.index, y=sales_by_product.values)
plt.title("Total Sales by Product Type")
plt.xlabel("Product Type")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

<span style="color:red; font-size:15px">2. Customer Segmentation by Age Group</span>


```python
bins = [0, 18, 25, 35, 45, 60, 100]
labels = ['0-18', '18-25', '25-35', '35-45', '45-60', '60+']
ecommerce_data['AgeGroup'] = pd.cut(ecommerce_data['CustomerAge'], bins, labels=labels)

# Plotting age distribution
plt.figure(figsize=(10, 7))
sns.countplot(data=ecommerce_data, x='AgeGroup', hue='CustomerGender')
plt.title("Customer Segmentation by Age Group and Gender")
plt.xlabel("Age Group")
plt.ylabel("Number of Customers")
plt.show()
```

<span style="color:red; font-size:15px">3. Product Performance</span>


```python
product_quantity = ecommerce_data.groupby('Product_Type')['Quantity'].sum()

# Plotting Quantity Sold per Product Type
plt.figure(figsize=(10, 8))
sns.barplot(x=product_quantity.index, y=product_quantity.values)
plt.title("Quantity Sold per Product Type")
plt.xlabel("Product Type")
plt.ylabel("Total Quantity Sold")
plt.xticks(rotation=45)
plt.show()
```

<span style="color:red; font-size:15px">4. Discount Effectiveness</span>


```python
# Calculating sales without discounts and applying discount factor
ecommerce_data['Discounted_Sales'] = ecommerce_data['Total_Price'] * (1 - ecommerce_data['Discount'])
ecommerce_data['Discount_Impact'] = ecommerce_data['Total_Price'] - ecommerce_data['Discounted_Sales']

# Plotting Discount Impact
plt.figure(figsize=(10, 5))
sns.histplot(ecommerce_data['Discount_Impact'], bins=30, kde=True)
plt.title("Distribution of Discount Impact on Sales")
plt.xlabel("Discount Impact")
plt.ylabel("Frequency")
plt.show()
```

<span style="color:red; font-size:15px">5. Payment Method Analysis</span>


```python
payment_method_sales = ecommerce_data.groupby('PaymentMethod')['Total_Sales'].sum()

# Plotting Sales by Payment Method
plt.figure(figsize=(8, 6))
sns.barplot(x=payment_method_sales.index, y=payment_method_sales.values)
plt.title("Total Sales by Payment Method")
plt.xlabel("Payment Method")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

<span style="color:red; font-size:15px">7. Customer Lifetime Value (CLV)</span>


```python
# Calculating total sales per customer
customer_lifetime_value = ecommerce_data.groupby('CustomerID')['Total_Sales'].sum()

# Plotting Customer Lifetime Value Distribution
plt.figure(figsize=(10, 5))
sns.histplot(customer_lifetime_value, bins=30, kde=True)
plt.title("Customer Lifetime Value Distribution")
plt.xlabel("Lifetime Value (Total Sales)")
plt.ylabel("Frequency")
plt.show()
```

<span style="color:red; font-size:15px">8. Geographical Analysis</span>


```python
import squarify


# Prepare data
sales_by_location = ecommerce_data.groupby('CustomerLocation')['Total_Sales'].sum()
locations = sales_by_location.index
sales = sales_by_location.values

# Plotting Sales by Location (Treemap)
plt.figure(figsize=(20, 8))
squarify.plot(sizes=sales, label=locations, value=sales, color=sns.color_palette("viridis", len(sales)), alpha=0.8)
plt.title("Sales by Customer Location (Treemap)")
plt.axis('off')  # Hide the axes for a clean look
plt.show()
```

<span style="color:red; font-size:15px">9. Average Order Value (AOV)</span>


```python
average_order_value = ecommerce_data['Total_Sales'].mean()
print(f"Average Order Value (AOV): {average_order_value:.2f}")
```

<span style="color:red; font-size:15px">10. Customer Purchase Frequency</span>


```python
purchase_frequency = ecommerce_data['CustomerID'].value_counts()

# Plotting Purchase Frequency Distribution
plt.figure(figsize=(10, 5))
sns.histplot(purchase_frequency, bins=30, kde=True)
plt.title("Customer Purchase Frequency Distribution")
plt.xlabel("Number of Purchases")
plt.ylabel("Frequency")
plt.show()
```
